```python
#Q1: what is the average interest rate for the bank?
fhb = loans.Bank("First Home Bank")
l_count = 0
sum_int_rate = 0
idx = 0
lfhb = len(fhb)
while lfhb > 0:
    lfhb -= 1
    loan = fhb[idx]
    idx += 1
    if loan.interest_rate > 0:
        sum_int_rate += loan.interest_rate
        l_count += 1

sum_int_rate / l_count
```

```python
#Q2: how many applicants are there per loan, on average?
lfhb = len(fhb)
l_count = 0
sum_a_count = 0
idx = 0

while lfhb > 0:
    lfhb -= 1
    loan = fhb[idx]
    idx += 1
    sum_a_count += len(loan.applicants)
    
sum_a_count / len(fhb)
```

```python
#Q3: what is the distribution of ages?
agedict = {}
for loan in fhb:
    for applicant in loan.applicants:
        if applicant.age not in agedict:
            agedict[applicant.age] = 1
        else:
            agedict[applicant.age] += 1
agedict
```

```python
#Q4: how many interest rate values are missing?
tfhb = search.BST()
for loan in fhb:
    tfhb.add(loan.interest_rate, loan)

len(tfhb[-1])
```

```python
#Q5: how tall is the tree?
tfhb.height(tfhb.root)
```

```python
#Q6: how long does it take to add the loans to the tree?
uwcu = loans.Bank("University of Wisconsin Credit Union")
uwcub = search.BST()
timelist = []
l_count = 0
stime = time.time()
for loan in uwcu:
    uwcub.add(loan.interest_rate, loan)
    etime = time.time()
    l_count += 1
    timelist.append(etime-stime)

plt.plot(range(l_count), timelist)
plt.xlabel('Loans')
plt.ylabel('Time Elapsed (Milliseconds)')
```

```python
#Q7: how fast are tree lookups?
t1s = time.time()
mcounter = 0
for loan in uwcu:
    if loan.interest_rate == -1:
        mcounter += 1
t1e = time.time()
time1 = t1e - t1s

t2s = time.time()
len(uwcub[-1])
t2e = time.time()
time2 = t2e - t2s

plt.bar(['Count Method Time', 'BST Method Time'], [time1, time2])
plt.ylabel('Time Elapsed (Milliseconds)')
```

```python
#Q8: How many applicants indicate multiple racial identities?
app_race_dict = {}
for loan in uwcu:
    for applicant in loan.applicants:
        if len(applicant.race) not in app_race_dict:
            app_race_dict[len(applicant.race)] = 1
        else:
            app_race_dict[len(applicant.race)] += 1

plt.bar(list(app_race_dict.keys()), list(app_race_dict.values()))
plt.xlabel('Number of Races Identified')
plt.ylabel('Number of Applicants')
```

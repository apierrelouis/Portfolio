```python
def smart_sum(nums=[0]):
    s = 0
    for n in nums:
        if n < 0:
            continue
        s += n
        if s >= 20:
            break
    return s
```

```python
#Q17: what are the keys in years?

surveyFiles = os.listdir(cwd + "/home-computers")
#print(surveyFiles)
files = []
for file in surveyFiles:
    if file[-4:] != ".csv":
        continue
    idx = file.find(".")
    year = file[idx-4:idx]
    files.append(int(year))

files.sort(key = int)
#print(files)
years = {}
for year in files:
    file = ""
    for idx in surveyFiles:
        if str(year) in idx:
            file = idx
            break
    years[year] = pandas.read_csv(cwd+"/home-computers/"+file,skiprows=1)
    years[year].set_index("Geographic Area Name",inplace=True)
#years[2019].index.values.tolist()
#years[2019].loc['Alabama','Geography']

list(years.keys())
```

```python
def rand_df(rows):
    return pandas.DataFrame(numpy.random.randint(10, size=(rows, 5)),
                        columns=["A", "B", "C", "D", "E"],
                        index=[f"r{i}" for i in range(1, rows+1)])

inp = 100
inputs = {}
while inp <= 5000:
    inputs[inp] = rand_df(inp)
    inp+=100
#print(inputs)
#ROWS DICT
inpTimes_iterrows = {}
for inp in inputs:
    t0 = time.time()
    for row in inputs[inp].iterrows():
        pass
    t1 = time.time()
    tpassed = t1-t0
    inpTimes_iterrows[inp] = tpassed
#TUPLES DICT
inpTimes_itertuples = {}
for inp in inputs:
    t0 = time.time()
    for row in inputs[inp].itertuples():
        pass
    t1 = time.time()
    tpassed = t1-t0
    inpTimes_itertuples[inp] = tpassed
```

```python
#Q30: what is faster for looping over a DataFrame, iterrows or itertuples?
#print(rand_df(5))
rowTimeSeries = pandas.Series(inpTimes_iterrows)
tupleTimeSeries = pandas.Series(inpTimes_itertuples)
iticks = list(inputs)
rowTimeSeries.plot(xticks=[1000,2000,3000,4000,5000])
tupleTimeSeries.plot()
plt.xlabel("Rows")
plt.ylabel("Milliseconds")
plt.title("Iterrows vs Itertuples Complexity")
plt.legend(["Iterrows","Itertuples"])
```

```python
#Q31: what is faster, loc, or at?
locTimes = {}
for inp in inputs:
    t0 = time.time()
    df = inputs[inp]
    total = 0
    for idx in df.index:
        for col in df.columns:
            total += df.loc[idx, col]
    t1 = time.time()
    tpassed = t1-t0
    locTimes[inp] = tpassed

atTimes = {}
for inp in inputs:
    t0 = time.time()
    df = inputs[inp]
    total = 0
    for idx in df.index:
        for col in df.columns:
            total += df.at[idx, col]
    t1 = time.time()
    tpassed = t1-t0
    atTimes[inp] = tpassed
    
locSeries = pandas.Series(locTimes)
atSeries = pandas.Series(atTimes)
locSeries.plot(xticks=[1000,2000,3000,4000,5000])
atSeries.plot()
plt.xlabel("Rows")
plt.ylabel("Milliseconds")
plt.title("loc vs at Complexity")
plt.legend(["loc","at"])
```

```python
#Q32: what is faster, a loop or .apply?
def laugh(x):
    return "qwerty" * x

rowIns = {}
inp = 500
while inp <= 20000:
    rowIns[inp] = rand_df(inp)
    inp += 500
#print(rowIns)
#Apply
applyTimes = {}
for inp in rowIns:
    t0 = time.time()
    df = rowIns[inp]
    result = df["A"].apply(laugh).tolist()
    t1 = time.time()
    tpassed = t1-t0
    applyTimes[inp] = tpassed
    
#Loop
loopTimes = {}
for inp in rowIns:
    t0 = time.time()
    df = rowIns[inp]
    result = []
    for val in df["A"]:
        result.append(laugh(val))
    t1 = time.time()
    tpassed = t1-t0
    loopTimes[inp] = tpassed
    
applySeries = pandas.Series(applyTimes)
loopSeries = pandas.Series(loopTimes)
applySeries.plot(xticks=[4000,8000,12000,16000,20000])
loopSeries.plot()
plt.xlabel("Rows")
plt.ylabel("Milliseconds")
plt.title(".apply vs loop Complexity")
plt.legend(["apply","loop"])
```
